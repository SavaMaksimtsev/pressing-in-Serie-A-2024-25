```python
import matplotlib.pyplot as plt
import pandas as pd

data = {
    'team': ['Аталанта', 'Интер', 'Лацио', 'Милан', 'Наполи', 'Фиорентина', 'Болонья', 
             'Ювентус', 'Рома', 'Удинезе', 'Кальяри', 'Венеция', 'Комо', 'Торино', 
             'Парма', 'Эмполи', 'Монца', 'Лечче', 'Геноа', 'Верона'],
    'opp_buildup': [82.33, 82.44, 82.24, 87.58, 84.79, 82.68, 77.09, 84.25, 81.52, 
                    83.81, 82.71, 82.73, 80.95, 84.35, 85.58, 86.09, 85.73, 
                    86.02, 84.29, 86.27],
    'opp_poss_lost': [23.80, 22.29, 21.00, 18.79, 17.79, 19.64, 25.71, 18.07, 
                      24.40, 22.36, 16.67, 20.53, 23.33, 22.80, 21.60, 19.13, 
                      17.57, 20.60, 24.40, 22.87]
}

df = pd.DataFrame(data)

plt.figure(figsize=(10, 6))
plt.scatter(df['opp_buildup'], df['opp_poss_lost'], color='#3a83c4')

como_data = df[df['team'] == 'Комо']
plt.scatter(como_data['opp_buildup'], como_data['opp_poss_lost'], color='#e05451', label='Como')

for i in range(len(df)):
    plt.annotate(df['team'][i], (df['opp_buildup'][i], df['opp_poss_lost'][i]), fontsize=8, ha='right')

plt.gca().invert_xaxis()
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

plt.xlabel('Точность передач соперника в билдапе')
plt.ylabel('Потери соперников в опасном радиусе')
plt.title('Эффективность прессинга в Серии А-2024/25')
plt.grid(False)

plt.tight_layout()
plt.savefig("como.png")
plt.show()
```
